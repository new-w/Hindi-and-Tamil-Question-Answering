### Wikipedia q&a dataset
`new_question`  
`new_answer_text`

```python
file = open("/content/WikiQA-train.txt", "r")

new_question = []
new_answer_text = []

while True:
    line = file.readline()
    sp_line = line.split('\t')
    n_q = sp_line[0]
    if len(sp_line) == 3:
        n_a = sp_line[1]
    if len(sp_line) == 3:
        label = sp_line[2]
    if label == '1\n':
        new_question.append(n_q)
        new_answer_text.append(n_a)

    if not line:
        break

for q, a in zip(new_question, new_answer_text):
    print("question: ", q)
    print("answer: ", a)
    print("\n")

print(len(new_question), len(new_answer_text))
file.close()
```

### Get context
`new_context`  
- question에 따른 answer를 예측하기 위해 answer_start가 필요
- answer_start를 얻기 위해서는 context가 필요
- Wikipedia 검색창에 question을 넣어서 결과 page가 정상적으로 나오면 해당 page를 이용
- 결과가 정상적으로 나오지 않을 경우 웹 크롤링을 통해 page를 얻어냄

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
b1 = "https://en.wikipedia.org/wiki/Special:Search?search="
b2 = "&go=Go&ns0=1"
def wiki_web(q_str):
    q_str = q_str.replace(' ', '+')
    url = b1 + q_str + b2

    html = urlopen(url)
    bs_html = BeautifulSoup(html, "html.parser")

    s = bs_html.select(".mw-search-result")
    s = str(s[0])
    s = s.split('href="')
    new_page = s[1].split(' ')[0]
    new_page = new_page[6:-1]
    return new_page
```

```python
import googletrans
translator = googletrans.Translator()
```

```python
for i in range(0, 100):
    if len(new_context[i]) > 5000:
        new_context[i] = new_context[i][0:5000]
```

```python
question_tamil = []
answer_text_tamil = []
context_tamil = []

for i in range(0, 100):
    question_tamil.append(translator.translate(new_question[i], dest = 'tamil').text)
    answer_text_tamil.append(translator.translate(new_answer_text[i], dest = 'tamil').text)
    if new_context[i] != '':
        print(i)
        context_tamil.append(translator.translate(new_context[i], dest = 'tamil').text)
    else:
        context_tamil.append('')
```

### Get answer_start
`new_answer_start`

```python
new_answer_start = []

def get_start(text, idx):
    target = text
    cnt = 0
    sig = 0
    start = 0
    for t in context_tamil[idx]:
        if t == target[cnt]:
            sig = 1
        else:
            sig = 0

        if sig == 1:
            cnt +=1
        else:
            cnt = 0
        start += 1
        # 원래는 10이 아니라 len(target)이 들어가는게 맞음
        if cnt == 10:
            print(idx, start - 10)
            new_answer_start.append((idx, start - 10))
            break
```

```python
for i in range(0, 100):
    get_start(question_tamil[i], i)
```
